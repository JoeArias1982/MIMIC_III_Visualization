```python
# List all files in the mimiciii-demo directory
import os

data_dir = './mimiciii-demo/'
files = os.listdir(data_dir)
print("Downloaded files:")
print(files)
```

```python
# Load each CSV file into a pandas DataFrame and store them in a dictionary
import pandas as pd

dataframes = {}  # Dictionary to store DataFrames

for file in files:
    if file.endswith('.csv'):  # Check if the file is a CSV
        file_path = os.path.join(data_dir, file)
        df_name = file.split('.')[0]  # Use the file name (without extension) as the key
        dataframes[df_name] = pd.read_csv(file_path)
        print(f"\nLoaded {file} into DataFrame with key '{df_name}'")

# Display all loaded DataFrames' keys
print("\nLoaded DataFrames:")
print(dataframes.keys())
```

```python
# Loop through each DataFrame in the dictionary and print its columns
for df_name, df in dataframes.items():
    print(f"Columns in DataFrame '{df_name}':")
    print(df.columns)
    print("\n")  # Add a newline for better readability between DataFrames
```

Number of Admissions Over Time

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Access the ADMISSIONS DataFrame
admissions_df = dataframes['ADMISSIONS']

# Convert the 'admittime' column to datetime
admissions_df['admittime'] = pd.to_datetime(admissions_df['admittime'])

# Extract the year and month from the admittime for time series analysis
admissions_df['admit_year_month'] = admissions_df['admittime'].dt.to_period('M')

# Count the number of admissions per month
admissions_per_month = admissions_df.groupby('admit_year_month').size()

# Plot the number of admissions over time
plt.figure(figsize=(12, 6))
admissions_per_month.plot(kind='line', color='blue')
plt.title('Number of Admissions Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Admissions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Length of Stay Distribution

```python
# Calculate the length of stay (LOS) in days
admissions_df['dischtime'] = pd.to_datetime(admissions_df['dischtime'])
admissions_df['los'] = (admissions_df['dischtime'] - admissions_df['admittime']).dt.days

# Plot the length of stay distribution
plt.figure(figsize=(10, 6))
sns.histplot(admissions_df['los'].dropna(), bins=30, kde=True, color='green')
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Admissions by Admission Location

```python
# Count the number of admissions by admission location
admission_location_counts = admissions_df['admission_location'].value_counts()

# Plot the number of admissions by location
plt.figure(figsize=(10, 6))
admission_location_counts.plot(kind='bar', color='purple')
plt.title('Number of Admissions by Admission Location')
plt.xlabel('Admission Location')
plt.ylabel('Number of Admissions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

 Insurance Distribution Among Patients

```python
# Count the number of patients by insurance type
insurance_counts = admissions_df['insurance'].value_counts()

# Plot the insurance distribution
plt.figure(figsize=(10, 6))
insurance_counts.plot(kind='bar', color='orange')
plt.title('Distribution of Patients by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Number of Procedures Performed Over Time (Using PROCEDUREEVENTS_MV)

```python
# Extracting the procedure start time
procedures_df = dataframes['PROCEDUREEVENTS_MV']
procedures_df['starttime'] = pd.to_datetime(procedures_df['starttime'])

# Count procedures per month
procedures_df['start_year_month'] = procedures_df['starttime'].dt.to_period('M')
procedures_per_month = procedures_df.groupby('start_year_month').size()

# Plotting procedures over time
plt.figure(figsize=(12, 6))
procedures_per_month.plot(kind='line', color='red')
plt.title('Number of Procedures Performed Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Procedures')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Mortality Rate by Admission Type (Using ADMISSIONS)

```python
# Calculating mortality rate by admission type
mortality_rate_by_type = admissions_df.groupby('admission_type')['hospital_expire_flag'].mean()

# Plotting the mortality rate by admission type
plt.figure(figsize=(10, 6))
mortality_rate_by_type.plot(kind='bar', color='red')
plt.title('Mortality Rate by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Mortality Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Correlation of Procedures and Lab Results

```python
# Access the necessary DataFrames
procedures_df = dataframes['PROCEDUREEVENTS_MV']
labevents_df = dataframes['LABEVENTS']

# Merge on 'hadm_id' (hospital admission ID)
merged_df = pd.merge(procedures_df, labevents_df, on='hadm_id', how='inner')

# Select numerical columns that exist in both DataFrames after merging
# We will focus on 'value' from both procedures and lab events
numerical_cols = ['value_x', 'value_y', 'valuenum']

# Drop any rows with missing values in the selected columns
correlation_df = merged_df[numerical_cols].dropna()

# Calculate the correlation matrix
corr_matrix = correlation_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Procedures and Lab Results')
plt.show()
```

Frequencies of Procedures by Location and Category

```python
# Group by 'location' and 'ordercategoryname'
procedure_location_category = procedures_df.groupby(['location', 'ordercategoryname']).size().unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(procedure_location_category, cmap='viridis', annot=False)
plt.title('Procedures Frequency by Location and Category')
plt.xlabel('Procedure Category')
plt.ylabel('Location')
plt.show()
```

Correlation Heatmap of ICU Stays and Mortality

```python
# Access the necessary DataFrames
icustays_df = dataframes['ICUSTAYS']
admissions_df = dataframes['ADMISSIONS']

# Merge on 'hadm_id' (hospital admission ID)
merged_df = pd.merge(icustays_df, admissions_df, on='hadm_id', how='inner')

# Print columns to verify what is available after the merge
print("Columns in merged DataFrame:")
print(merged_df.columns)

# Adjust correlation_cols based on what is actually available
correlation_cols = ['hospital_expire_flag', 'icustay_id', 'first_wardid', 'last_wardid']

# Filter only the columns that exist
available_cols = [col for col in correlation_cols if col in merged_df.columns]

# Drop any rows with missing values in the selected columns
correlation_df = merged_df[available_cols].dropna()

# Calculate the correlation matrix
corr_matrix = correlation_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='magma', fmt='.2f')
plt.title('Correlation Heatmap of ICU Stays and Mortality')
plt.show()
```

```python
# Merge ADMISSIONS with PATIENTS to get DOB and Gender
admissions_df = dataframes['ADMISSIONS']
patients_df = dataframes['PATIENTS']
merged_df = pd.merge(admissions_df, patients_df, on='subject_id', how='inner')

# Convert 'admittime' and 'dob' to datetime, with errors handled by setting invalid parsing to NaT
merged_df['admittime'] = pd.to_datetime(merged_df['admittime'], errors='coerce')
merged_df['dob'] = pd.to_datetime(merged_df['dob'], errors='coerce')

# Print out the number of rows before filtering
print(f"Total rows before filtering: {len(merged_df)}")

# Filter out rows with invalid or extreme dates
valid_dates_df = merged_df.dropna(subset=['admittime', 'dob'])
valid_dates_df = valid_dates_df[(valid_dates_df['admittime'] >= valid_dates_df['dob']) &
                                (valid_dates_df['dob'] > '1900-01-01') &  # Filter for reasonable DOBs
                                (valid_dates_df['admittime'] < '2100-01-01')]  # Ensure no future dates

# Print out the number of rows after filtering
print(f"Total rows after filtering: {len(valid_dates_df)}")

# If there's still data, proceed with calculating age
if not valid_dates_df.empty:
    # Calculate the age at admission safely
    valid_dates_df['age'] = ((valid_dates_df['admittime'] - valid_dates_df['dob']).dt.days // 365).astype(int)

    # Scatter plot of Age vs. Length of Stay by Gender
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='age', y='los', hue='gender', data=valid_dates_df)
    plt.title('Age vs Length of Stay by Gender')
    plt.xlabel('Age')
    plt.ylabel('Length of Stay (Days)')
    plt.grid(True)
    plt.show()
else:
    print("No data available after filtering.")
```

```python
# Relaxed DOB filter and adjusted criteria
valid_dates_df = valid_dates_df[(valid_dates_df['admittime'] >= valid_dates_df['dob']) &
                                (valid_dates_df['dob'] > '1800-01-01') &  # Adjust DOB filter
                                (valid_dates_df['admittime'] < '2100-01-01')]

# Calculate the age at admission safely
valid_dates_df['age'] = ((valid_dates_df['admittime'] - valid_dates_df['dob']).dt.days // 365).astype(int)

# Scatter plot of Age vs. Length of Stay by Gender
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='los', hue='gender', data=valid_dates_df)
plt.title('Age vs Length of Stay by Gender')
plt.xlabel('Age')
plt.ylabel('Length of Stay (Days)')
plt.grid(True)
plt.show()
```

Lab Values by Diagnosis

```python
# Access the necessary DataFrames
# LABEVENTS contains lab test results, DIAGNOSES_ICD contains patient diagnoses,
# D_LABITEMS contains lab test names and descriptions, D_ICD_DIAGNOSES contains ICD code descriptions
labevents_df = dataframes['LABEVENTS']
diagnoses_df = dataframes['DIAGNOSES_ICD']
labitems_df = dataframes['D_LABITEMS']
icd_diag_df = dataframes['D_ICD_DIAGNOSES']

# Merge LABEVENTS with DIAGNOSES_ICD on 'hadm_id' to link lab test results with patient diagnoses
merged_df = pd.merge(labevents_df, diagnoses_df, on='hadm_id', how='inner')

# Prompt the user to input the lab test ID (itemid)
# This allows the user to select which lab test they want to analyze
try:
    specific_lab_id = int(input("Enter the itemid for the lab test you want to analyze: "))

    # Get the name of the lab test associated with the entered itemid from D_LABITEMS
    # If the lab test name is not found, default to "Unknown Lab Test"
    lab_name = labitems_df.loc[labitems_df['itemid'] == specific_lab_id, 'label'].values
    lab_name = lab_name[0] if len(lab_name) > 0 else "Unknown Lab Test"

    # Filter the merged DataFrame to include only rows corresponding to the specific lab test
    specific_lab = merged_df[merged_df['itemid'] == specific_lab_id]

    # Merge the filtered data with ICD-9 descriptions to get full diagnosis names
    specific_lab = pd.merge(specific_lab, icd_diag_df, on='icd9_code', how='left')

    # Limit the dataset to the top 10 most common ICD-9 diagnoses for better readability
    top_icd_codes = specific_lab['icd9_code'].value_counts().head(10).index
    specific_lab = specific_lab[specific_lab['icd9_code'].isin(top_icd_codes)]

    # Ensure there is data to plot; if not, inform the user
    if not specific_lab.empty:
        # Increase the figure size to make the plot more readable
        plt.figure(figsize=(14, 8))

        # Create a violin plot to show the distribution of lab values by diagnosis
        # X-axis: ICD-9 diagnosis descriptions (long_title)
        # Y-axis: Lab test values (valuenum)
        sns.violinplot(x='long_title', y='valuenum', data=specific_lab)

        # Title includes both the lab test name and ID for clarity
        plt.title(f'Distribution of {lab_name} (Lab Test ID: {specific_lab_id}) by Top Diagnoses')

        # Label the x and y axes
        plt.xlabel('ICD-9 Diagnosis')
        plt.ylabel('Lab Test Value')

        # Rotate x-axis labels to ensure they are readable and do not overlap
        plt.xticks(rotation=45, ha='right')

        # Use tight_layout to adjust the layout and ensure all labels fit within the figure
        plt.tight_layout()

        # Display the plot
        plt.show()
    else:
        # If no data is available for the specified lab test ID, notify the user
        print(f"No data available for the specified lab test ID: {specific_lab_id}")
except ValueError:
    # Handle cases where the user input is not a valid number
    print("Invalid input. Please enter a numeric lab test ID.")
```
